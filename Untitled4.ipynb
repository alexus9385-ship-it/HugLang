### 1. Добавление функции сохранения контрольной точки (checkpoint)

Вам понадобится функция, которая будет сохранять состояние вашей модели, оптимизатора, текущую эпоху и, возможно, лучшую достигнутую точность. Вы можете добавить этот код в `main.py` или в отдельный утилитарный файл, который импортируется в `main.py`.

**Пример функции сохранения (сохраняет модель при каждом новом "лучшем" результате):**

````python
import torch
import os

def save_checkpoint(state, is_best, checkpoint_dir='checkpoints', filename='checkpoint.pth.tar'):
    """
    Сохраняет контрольную точку (checkpoint) обучения.
    Args:
        state (dict): Словарь, содержащий состояние модели, оптимизатора, эпоху и т.д.
        is_best (bool): Флаг, указывающий, является ли текущая модель лучшей.
        checkpoint_dir (str): Директория для сохранения контрольных точек.
        filename (str): Имя файла для сохранения контрольной точки.
    """
    if not os.path.exists(checkpoint_dir):
        os.makedirs(checkpoint_dir)
    filepath = os.path.join(checkpoint_dir, filename)
    torch.save(state, filepath)
    if is_best:
        # Сохраняем также отдельный файл для лучшей модели
        best_filepath = os.path.join(checkpoint_dir, 'model_best.pth.tar')
        torch.save(state, best_filepath)
        print(f"⭐ NEW BEST MODEL SAVED to {best_filepath}")


# Пример использования в вашем цикле обучения:
# В вашем training loop (вероятно, в функции `train` или `main` в `main.py`):
# ...
# внутри цикла по эпохам:
# for epoch in range(start_epoch, num_epochs):
#     # ... ваш код обучения и валидации ...
#
#     # После валидации, получаем текущую точность или метрику
#     current_accuracy = validation_accuracy # Предположим, вы вычисляете это
#     is_best = current_accuracy > best_accuracy
#     best_accuracy = max(current_accuracy, best_accuracy)
#
#     save_checkpoint({
#         'epoch': epoch + 1,
#         'arch': 'your_model_architecture_name',
#         'state_dict': model.state_dict(),
#         'best_accuracy': best_accuracy,
#         'optimizer': optimizer.state_dict(),
#         'lr_scheduler': lr_scheduler.state_dict() if lr_scheduler else None,
#     }, is_best)
# ...

print("Это пример функции `save_checkpoint`. Вам нужно будет интегрировать её в ваш `main.py`.")
````

### 2. Добавление функции для загрузки контрольной точки и продолжения обучения

Эта функция позволит вам загрузить сохраненную контрольную точку и инициализировать модель, оптимизатор, планировщик скорости обучения и продолжить с нужной эпохи.

````python
import torch
import os

def load_checkpoint(filepath, model, optimizer=None, lr_scheduler=None):
    """
    Загружает контрольную точку обучения.
    Args:
        filepath (str): Путь к файлу контрольной точки.
        model (torch.nn.Module): Модель PyTorch.
        optimizer (torch.optim.Optimizer, optional): Оптимизатор PyTorch.
        lr_scheduler (torch.optim.lr_scheduler._LRScheduler, optional): Планировщик скорости обучения PyTorch.
    Returns:
        tuple: (model, optimizer, lr_scheduler, start_epoch, best_accuracy) или None, если файл не найден.
    """
    if os.path.isfile(filepath):
        print(f"=> Loading checkpoint '{filepath}'")
        checkpoint = torch.load(filepath, map_location='cuda' if torch.cuda.is_available() else 'cpu')
        start_epoch = checkpoint['epoch']
        best_accuracy = checkpoint['best_accuracy']
        model.load_state_dict(checkpoint['state_dict'])
        if optimizer and 'optimizer' in checkpoint:
            optimizer.load_state_dict(checkpoint['optimizer'])
        if lr_scheduler and 'lr_scheduler' in checkpoint and checkpoint['lr_scheduler'] is not None:
            lr_scheduler.load_state_dict(checkpoint['lr_scheduler'])
        print(f"=> Loaded checkpoint '{filepath}' (epoch {checkpoint['epoch']} with best accuracy {checkpoint['best_accuracy']:.2f}%) ")
        return model, optimizer, lr_scheduler, start_epoch, best_accuracy
    else:
        print(f"=> No checkpoint found at '{filepath}'")
        return None


# Пример использования при запуске обучения в вашем main.py:
#
# # В начале вашего main.py или функции, которая запускает обучение:
# model = YourModelArchitecture(...)
# optimizer = YourOptimizer(model.parameters(), ...)
# lr_scheduler = YourLRScheduler(optimizer, ...)
#
# start_epoch = 0
# best_accuracy = 0.0
#
# checkpoint_path = 'checkpoints/checkpoint.pth.tar' # Или 'checkpoints/model_best.pth.tar'
#
# loaded_state = load_checkpoint(checkpoint_path, model, optimizer, lr_scheduler)
# if loaded_state:
#     model, optimizer, lr_scheduler, start_epoch, best_accuracy = loaded_state
#     print(f"Continuing training from epoch {start_epoch}, with best accuracy {best_accuracy:.2f}%")
# else:
#     print("Starting training from scratch.")
#
# # Теперь ваш цикл обучения начнется с `start_epoch`
# for epoch in range(start_epoch, num_epochs):
#     # ... ваш код обучения ...

print("Это пример функции `load_checkpoint`. Вам нужно будет интегрировать её в ваш `main.py`.")
````

### Важное замечание по `lr_scheduler`:

Ранее я уже упоминал о возможном некорректном порядке вызова `lr_scheduler.step()` и `optimizer.step()`. Убедитесь, что в вашем цикле обучения `optimizer.step()` вызывается *перед* `lr_scheduler.step()` для корректного обновления скорости обучения.

Пример корректного порядка:

```python
# В конце каждой итерации или эпохи:
optimizer.step()
lr_scheduler.step()
```
